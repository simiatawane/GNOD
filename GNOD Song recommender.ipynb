# List of top 100 songs from popvortex

```python
# 1. find url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php/"
```

```python
# 2. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 3. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4. check that the html code looks like it should
#soup
```

```python
title = []
for a in soup.select(".title"):
    title.append(a.get_text())
```

```python
title
```

```python
artist = []
for a in soup.select(".artist"):
    artist.append(a.get_text())
```

```python
artist
```

```python
songs_top_100 = pd.DataFrame({"title":title,
                       "artist":artist,
                      })
```

```python
songs_top_100
```

```python
songs_top_100.index = songs_top_100.index + 1
```

# Creating a song recommender

```python
#Thanks to chatgpt 

import pandas as pd

data = songs_top_500
df = pd.DataFrame(data)

song_name = input("Please enter the name of a song: ")

if any(c.isalpha() for c in song_name) and any(c.isalnum() for c in song_name):
    song_name = song_name.lower()

    matching_song = df[(df['titles'].str.lower() == song_name)]

    if not matching_song.empty:
        print(f"The song '{matching_song['titles'].values[0]}' by {matching_song['artists'].values[0]} matches your input.")

        recommended_song = df[df['titles'].str.lower() != song_name].sample(1)
        print(f"Here's a recommended song: '{recommended_song['titles'].values[0]}' by {recommended_song['artists'].values[0]}")
    else:
        print(f"No match found for the song: {song_name}")
else:
    print("Invalid input. Please enter a song name containing letters and/or numbers.")
```
